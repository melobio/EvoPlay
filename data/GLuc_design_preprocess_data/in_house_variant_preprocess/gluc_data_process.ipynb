# data preprocess and fitness alignment

```python
import pandas as pd
import numpy as np
import re

gluc_wt = "KPTENNEDFNIVAVASNFATTDLDADRGKLPGKKLPLEVLKEMEANARKAGCTRGCLICLSHIKCTPKMKKFIPGRCHTYEGDKESAQGGIGEAIVDIPEIPGFKDLEPMEQFIAQVDLCVDCTTGCLKGLANVQCSDLLKKWLPQRCATFASKIQGQVDKIKGAGGD"
mutation_distribution = np.zeros(168)
print("wt长度{}".format(len(gluc_wt)))
print(gluc_wt[25],gluc_wt[31],gluc_wt[161])
key_list = []
fit_list = []
sheet_list = []
#以gluc3为基准进行标准化
df_3 = pd.read_excel(r"D:\数据\Gluc荧光素酶\5-Gluc突变体位点信息.xlsx",sheet_name="Gluc-3-蛋白-（B6基础上）特异性突变体")
dict_3 = {}
#print(df_3["突变位点"])
#print(df_3.iloc[2]["突变位点"].split(", "))
for i in range(len(df_3)-1):
    mutants = df_3.iloc[i]["突变位点"].split(",")
    mutants = [mut.strip() for mut in mutants]
    #print(mutants)
    gluc_wt_tmp = list(gluc_wt)
    #print(gluc_wt_tmp)
    for mut in mutants:
        #print(int(mut[1:-1]))
        try:
            position = int(mut[1:-1])-1
            mutation_distribution[position] += 1
        except:
            print("error_{}_{}".format(i,mut[1:-1]))
        assert mut[0]==gluc_wt_tmp[position]
        gluc_wt_tmp[position] = mut[-1]
    gluc_wt_tmp = "".join(gluc_wt_tmp)
    print(gluc_wt_tmp)
    fit = df_3.iloc[i]["fitness"]
    #print(type(fit))
    # if type(fit) == float:
    #     print(type(fit))
    dict_3[gluc_wt_tmp] = fit

df_2 = pd.read_excel(r"D:\数据\Gluc荧光素酶\5-Gluc突变体位点信息.xlsx",sheet_name="Gluc-2-蛋白-活性较低的突变体")
dict_2 = {}
for i in range(len(df_2)):
    mutants = df_2.iloc[i]["突变体相较于WT进行突变的位点"].split(" ")
    mutants = [mut.strip() for mut in mutants if mut!=""]
    #print(mutants)
    gluc_wt_tmp = list(gluc_wt)
    #print(gluc_wt_tmp)
    for mut in mutants:
        print(mut[1:-1])
        try:
            position = int(mut[1:-1])-1
            mutation_distribution[position] += 1
        except:
            print(i)
            print("error_{}_{}".format(i,mutants))
        try:
            assert mut[0]==gluc_wt_tmp[position]
        except:
            print("第{}行".format(i),mut[0],gluc_wt_tmp[position])
        gluc_wt_tmp[position] = mut[-1]
    gluc_wt_tmp = "".join(gluc_wt_tmp)
    print(gluc_wt_tmp)
    fit = df_2.iloc[i]["fitness"]
    #if type(fit) == float:
    dict_2[gluc_wt_tmp] = fit

df_1 = pd.read_excel(r"D:\数据\Gluc荧光素酶\5-Gluc突变体位点信息.xlsx",sheet_name="Gluc-1-蛋白-活性突变体")
dict_1 = {}
for i in range(len(df_1)-2):
    mutants = df_1.iloc[i]["突变"].split(",")
    mutants = [mut.strip() for mut in mutants if mut!=""]
    #print(mutants)
    gluc_wt_tmp = list(gluc_wt)
    #print(gluc_wt_tmp)
    for mut in mutants:
        print(mut[1:-1])
        try:
            position = int(mut[1:-1])-1
            mutation_distribution[position] += 1
        except:
            print(i)
            print("error_{}_{}".format(i,mutants))
        try:
            assert mut[0]==gluc_wt_tmp[position]
        except:
            print("第{}行".format(i),mut[0],gluc_wt_tmp[position])
        gluc_wt_tmp[position] = mut[-1]
    gluc_wt_tmp = "".join(gluc_wt_tmp)
    print(gluc_wt_tmp)
    fit = df_1.iloc[i]["fitness"]
    #if type(fit) == float:
    dict_1[gluc_wt_tmp] = (fit*3.052)/4.74

df_4 = pd.read_excel(r"D:\数据\Gluc荧光素酶\5-Gluc突变体位点信息.xlsx",sheet_name="Gluc-4蛋白-B6基础上-2")
dict_4 = {}
for i in range(2,len(df_4)):
    mutants = df_4.iloc[i]["突变"].split(",")
    mutants = [mut.strip() for mut in mutants if mut!=""]
    #print(mutants)
    gluc_wt_tmp = list(gluc_wt)
    #print(gluc_wt_tmp)
    for mut in mutants:
        print(mut[1:-1])
        try:
            position = int(mut[1:-1])-1
            mutation_distribution[position] += 1
        except:
            print(i)
            print("error_{}_{}".format(i,mutants))
        try:
            assert mut[0]==gluc_wt_tmp[position]
        except:
            print("第{}行".format(i),mut[0],gluc_wt_tmp[position])
        gluc_wt_tmp[position] = mut[-1]
    gluc_wt_tmp = "".join(gluc_wt_tmp)
    print(gluc_wt_tmp)
    fit = df_4.iloc[i]["fitness"]
    #if type(fit) == float:
    dict_4[gluc_wt_tmp] = (fit*3.052)/2.6
#
key_list.extend(list(dict_1.keys()))
key_list.extend(list(dict_2.keys()))
key_list.extend(list(dict_3.keys()))
key_list.extend(list(dict_4.keys()))

fit_list.extend(list(dict_1.values()))
fit_list.extend(list(dict_2.values()))
fit_list.extend(list(dict_3.values()))
fit_list.extend(list(dict_4.values()))

sheet_list.extend(["sheet1"]*len(dict_1))
sheet_list.extend(["sheet2"]*len(dict_2))
sheet_list.extend(["sheet3"]*len(dict_3))
sheet_list.extend(["sheet4"]*len(dict_4))

ndf = pd.DataFrame({"mutants":key_list,"fitness_relative":fit_list, "sheet":sheet_list})
ndf.to_csv(r"D:\数据\Gluc荧光素酶\data_process_demo\gluc_fitness_relative.csv", index=False)
ddf = pd.DataFrame({"mutants":list(gluc_wt),"mutation_times":mutation_distribution.tolist()})
ddf.to_csv(r"D:\数据\Gluc荧光素酶\data_process_demo\gluc_mutation_distribution.csv", index=False)
```
